## The basics

We start by importing matplotlib, notice we import not `matplotlib` but `pyplot` and we use an alias as is commonly used out there. We will import `numpy` as well. Matplotlib uses NumPy's array structure, and at the same time we will use it to generate some data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

There are many ways to do a chart (_plot_ as it is known in Matplotlib), for these examples we will use the _direct_ way, this is a global Plot object where we just instruct the global `plt` module what we want to plot:

```python
data = np.arange(0, 5, 0.2)
plt.plot(data)
```

**NOTE**: For now we will be using a simple linear chart, in Pyplot is known as _plot_, we will see different types later when we know a little more about Matplotlib, but for now, `plot` is a good way to see our data.

In `pyplot` there are many ways to specify the data, if you pass a simple array it will use its values as the `y` axis values (as we used in the previous example), if we pass two arrays or lists the first will represent the `x` axis and the second the `y` axis:

```python
plt.plot(data, np.arange(5, 10, .2))
```

We can continue passing more than two lists or arrays and those will be taken as an additional serie in the plot, for example:

```python
plt.plot(data, data * 2, data ** 3, data / 2, data **2)
```

Another option to plot different series in the same plot is using multiple plot commands _before_ issuing the `show` command (which is automatically issued when you press enter in Jupyter and have the `%matplot inline` set)

```python
data = np.arange(0, 5, .2)
plt.plot(data, data)
plt.plot(data, data * 2)
plt.plot(data, data ** 2)
```

When we pass series it will assume we are passing them _in pairs_, so passing `a, b, c` will be assumed as `x1, y1, y2` while `a, b, c, d` is `x1, y1, x2, y2`. Yes, this is extremely confussing and it took me a while to understand. **Add note about how is done in ggplot2 as an example**

We can map cathegorical data as well, the same rules of `x` and `y` axis applies

```python
plt.plot(['Garfield', 'Eys', 'Silvester'], [12, 7, 4])
```

## Line properties

If confussion about passing series to methods like `plot` is not enough, you can pass a _string_ specifying the line style for your plot, for example:

```python
plt.plot(np.arange(10, 20), '--')
```

There are a lot of different codes and styles documented in [Plot documentation, at the notes](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot), you can specify colors and style all together:

```python
plt.plot(np.arange(10, 20), 'r--')
```

While this is a _nice shortcut_, I won't recommend at all using it mostly because it is confussing for anybody looking at your code, instead, prefer using the keyword parameters specifying any of the properties for the [Line2D](https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D) object:

```python
plt.plot(np.arange(10, 20), color='r', linestyle='--')
```

See? a lot clearer!

Let's take another example, let's say you want to use a marker instead of a continous line, the equivalent using keyword arguments will matter of just simple checking the required parameters of Line2D and passing them to the plot:

```python
plt.plot(np.arange(10, 20), color='r', marker='o', linestyle='')
```

While doing the same using the string format will be a "little more cryptic":

```python
plt.plot(np.arange(10, 20), 'ro ')
```

What if you want to plot a function instead of a specific set of data? Well, there are a few ways to do this, one is using a vectorize function from NumPy. There are a good reference about vectorized functions in NumPy in this very useful tutorial [From Python to NumPy](https://www.labri.fr/perso/nrougier/from-python-to-numpy/):

```python
x = np.arange(-4, 10, .1)
y = np.sin(x)

plt.plot(x, y)
```
